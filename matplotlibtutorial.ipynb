## Line Chart

```python
import numpy as np
import matplotlib.pyplot as plt 
# Matplotlib is used to visualize data
```

```python
x = np.linspace(0,20,1000)  # 1D array with 1000 evenly spaced points between 0 and 20
```

```python
y = np.sin(x) + 0.2*x
```

```python
plt.plot(x,y)  # Doesn't show if you are not using a notebook
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('input')
plt.ylabel('output')
plt.title('my plot');
```

## Scatterplot

```python
X = np.random.randn(100, 2)
plt.scatter(X[:,0], X[:,1])
```

```python
X = np.random.randn(200, 2)
X[:50] += 3
Y = np.zeros(200)
Y[:50] = 1
plt.scatter(X[:,0], X[:,1], c= Y);  # c is one dimensional array via Y
```

```python
from PIL import Image # Pillow, or Python Imaging Library
im = Image.open('Stanley_Park_Labeled_Aerial_Map.png')
```

```python
type(im)
```

```python
import numpy as np
arr = np.array(im)  # Works because images are represented as arrays of pixels

```

```python
arr  # All numbers between 0 and 255, representing the intensity of the colour at each pixel
```

```python
arr.shape  # First 2 dimensions are the size of the image, the third is the colour channels (RGB)
```

```python
from matplotlib import pyplot as plt
plt.imshow(arr)  # Plot the image
```

```python
gray = arr.mean(axis=2)  # Average the RGB channels to get a grayscale image
```

```python
gray.shape
```

```python
plt.imshow(gray); 
```

```python
plt.imshow(gray, cmap='gray');  # Plot the image in grayscale
```

```python
plt.imshow(arr, cmap='gray');  # Doesn't work because the image is not grayscale
```
