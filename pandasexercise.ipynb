```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

"""
Pandas Exercise
Generate, plot, and save "donut (two concentric circles) dataset" with quadratic features
Create a dataframe with just (x1,x2,y) first, the original coordinates
Then add x1^2, x2^2, x1*x2, the feature expansion
Save to csv without header row, index column

"""

# Generate donut dataset
N = 1000
# Array of 1000 random numbers between 0 and 2pi
theta = 2 * np.pi * np.random.rand(N)  
# Radius of 1 + some random noise
radius1 = 5 + 1 * np.random.randn(N)
radius2 = 10 + 1 * np.random.randn(N)
# x and y coordinates of inner circle
x1_inner = radius1 * np.cos(theta)
x2_inner = radius1 * np.sin(theta)
#  x and y coordinates of outer circle
x1_outer = radius2 * np.cos(theta)
x2_outer = radius2 * np.sin(theta)
# Concatenate inner and outer circle coordinates
x1 = np.concatenate([x1_inner, x1_outer])
x2 = np.concatenate([x2_inner, x2_outer])
#  0 for inner circle, 1 for outer circle
y = np.array([0]*N + [1]*N)

# Create dataframe
data = pd.DataFrame({'x1': x1, 'x2': x2, 'y': y})
data['x1^2'] = data['x1']**2
data['x2^2'] = data['x2']**2
data['x1*x2'] = data['x1']*data['x2']

# Save to CSV
data.to_csv('donut_dataset.csv', index=False, header=False, columns=['x1', 'x2', 'y', 'x1^2', 'x2^2', 'x1*x2'])

# Plot dataset
plt.scatter(x1_inner, x2_inner, color='purple', alpha=0.8)  #
plt.scatter(x1_outer, x2_outer, color='yellow', alpha=0.8)
plt.xlim(-15, 15)
plt.ylim(-15, 15)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Pandas Exercise: Donuts')
plt.show()
```
