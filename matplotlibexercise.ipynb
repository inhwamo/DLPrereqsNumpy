```python
# Exercise: generate and plot the XOR dataset
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Scatter plot with scales -1 to 1 on both axes
X = np.random.rand(800, 2)  # 400 random points in 2D
# 200 points in the lower left quadrant
X[:200] -= 1 
# 200 points in the lower right quadrant
X[200:400] = X[200:400] - [0, 1] 
# 200 points in the upper left quadrant
X[400:600] = X[400:600] - [1, 0]
# No need to change the upper right quadrant
Y = np.zeros(800)
Y[:200] = -1
Y[600:] = -1
plt.scatter(X[:, 0], X[:, 1], c= Y)  # c is the color
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.show()

```
