## PDF and CDF

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
x = np.linspace(-6,6,1000) 
```

```python
fx = norm.pdf(x,loc=0,scale=1)  # loc = mean, scale = standard deviation
```

```python
plt.plot(x,fx);
```

```python
Fx = norm.cdf(x,loc=0,scale=1)
```

```python
plt.plot(x,Fx);
```

```python
logfx = norm.logpdf(x,loc=0,scale=1)
```

```python
plt.plot(x,logfx);
```

## Convolution

```python
from PIL import Image
# Convolution is a common operation in image processing, and deep learning, etc
```

```python
!wget https://upload.wikimedia.org/wikipedia/commons/7/76/Stanley_Park_Labeled_Aerial_Map.png
```

```python
im = Image.open('Stanley_Park_Labeled_Aerial_Map.png')
```

```python
gray = np.mean(im,axis=2)
```

```python
x = np.linspace(-6,6,50)
```

```python
fx = norm.pdf(x,loc=0,scale=1)
```

```python
filt = np.outer(fx, fx)  # Gaussian filter
```

```python
plt.imshow(filt,cmap='gray');
```

```python
from scipy.signal import convolve2d
```

```python
out = convolve2d(gray,filt)
```

```python
plt.subplot(1,2,1)  # (rows,columns,plot number)
plt.imshow(gray,cmap='gray')
plt.subplot(1,2,2)
# Used convolution to blur the image using a Gaussian filter
plt.imshow(out,cmap='gray');
```
