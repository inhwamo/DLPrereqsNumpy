```python
import numpy as np
import matplotlib.pyplot as plt 
# Matplotlib is used to visualize data
```

```python
x = np.linspace(0,20,1000)  # 1D array with 1000 evenly spaced points between 0 and 20
```

```python
y = np.sin(x) + 0.2*x
```

```python
plt.plot(x,y)  # Doesn't show if you are not using a notebook
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('input')
plt.ylabel('output')
plt.title('my plot');
```
